```python
import pandas as pd
df = pd.read_csv('./data/iris.data', names=['sepal length', 'sepal width', 'pedal length', 'pedal width', 'fclass'])
```

```python
df.describe()
```

```python
set(df.fclass)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4], df.iloc[:, -1], test_size=0.25)
```

```python
colors = {'Iris-setosa': 'red', 'Iris-versicolor': 'blue', 'Iris-virginica': 'green'}
#scat_mat = pd.plotting.scatter_matrix(df.iloc[:, 0:4], c=df.iloc[:, -1].apply(lambda x: colors[x]), figsize=(15, 15), marker='o', hist_kwds={'bins': 20}, s=60, alpha=0.7)
scat_mat = pd.plotting.scatter_matrix(X_train, c=y_train.apply(lambda x: colors[x]), figsize=(10, 10), marker='o', hist_kwds={'bins': 20}, s=60, alpha=0.7)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

```python
import numpy as np
X_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(X_new)
print(prediction)
```

```python
prediction = knn.predict(X_test)
print(prediction)
```

```python
knn.score(X_test, y_test)
```
